# 4. 데이터 가공 1 - 데이터 통합
## 4-1. pandas.merge()

```python
#샘플 데이터 생성
df1 = DataFrame([['홍길동',23,'남'],['이순신',55,'남'], ['신사임당',48,'여']], 
                           columns = ['이름','나이','성별'])
df2 = DataFrame([['홍길동','터미네이터',5],['홍길동','비긴어게인',2], 
                 ['홍길동','살인의추억',4], ['신사임당','비긴어게인',5],
                ['장보고','살인의추억',3]], 
                           columns = ['이름','영화','평점'])
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
# df1.merge(df2)
```

```python
pd.merge(df1, df2, on='이름', how='inner')
```

## 실제 데이터 실습
* data/movielens 에 있는 영화 평점 데이터들을 하나로 합치기

```python
users = pd.read_csv('data/movielens/users.dat', sep='::', engine='python',
                   names=['사용자아이디','성별','나이','직업','지역'])
users.head()
```

```python
ratings = pd.read_csv('data/movielens/ratings.dat', sep='::', engine='python',
                     names=['사용자아이디','영화아이디','평점','시간'])
ratings.head()
```

```python
movies = pd.read_csv('data/movielens/movies.dat', sep='::', engine='python',
                    names=['영화아이디','제목','장르'])
movies.head()
```

```python
data=pd.merge(users, ratings)
data
```

```python
data=pd.merge(data, movies)
data
```

## 4-2. pandas.concat(axis = 0)
* 동일한 컬럼들을 가지고 있는 데이터프레임들을 하나로 합치고 싶은 경우

```python
df1 = DataFrame([['홍길동',23,'남'],['이순신',55,'남']], 
                           columns = ['이름','나이','성별'])
df2 = DataFrame([['신사임당',48,'여'], ['장보고',22,'남']], 
                           columns = ['이름','나이','성별'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2], axis=0)
```

```python
df1 = DataFrame([['홍길동',23,'남'],['이순신',55,'남']], 
                           columns = ['이름','나이','성별'])
df2 = DataFrame([['신사임당',48,'여'], ['장보고',22,'남']], 
                           columns = ['이름','나이','성별'])
pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
df1 = DataFrame([['홍길동',23,'남'],['이순신',55,'남']], 
                           columns = ['이름','나이','성별'])
df2 = DataFrame([['신사임당',48,'여'], ['장보고',22,'남']], 
                           columns = ['이름','나이','성별2'])
pd.concat([df1, df2], axis=0)
```

## 4-3. concat(axis = 1)
* 서로 다른 컬럼을 가진 데이터프레임들을 로우 인덱스의 값을 기준으로 합치고 싶은 경우

```python
# 샘플 데이터
df1 = DataFrame([[23,'남'],[55,'남'], [48,'여']], 
                           columns = ['나이','성별'], index = ['홍길동','이순신','신사임당'])
df2 = DataFrame([['서울',3], ['경기', 1]], 
                           columns = ['주소','학년'], index = ['홍길동','신사임당'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], axis=1, join='inner')
```
