```python
from keras.datasets import imdb
from keras.preprocessing import sequence

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=10000)
X_train = sequence.pad_sequences(X_train, maxlen=50)
X_test = sequence.pad_sequences(X_test, maxlen=50)
```

```python
data = imdb.get_word_index()
```

```python
data2 = {}
for key, value in data.items():
    data2[value+3] = key # data 내에서는 the->1, data2에서는 4->the
```

```python
for idx, word in enumerate(('<pad', '<sos>', '<unk>')):
    data2[idx] = word
```

```python
for j in range(5):
    print(' '.join([data2[i] for i in X_train[j]]))
    print()
```

```python
for j in range(5):
    print(y_train[j])
```

```python
from keras.layers.embeddings import Embedding
from keras.models import Sequential
from keras.layers import LSTM, Dense

model = Sequential()
model.add(Embedding(input_dim = 10000,
                    output_dim = 32,
                    input_length = 50))
model.add(LSTM(units=100))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train,
          epochs=10, batch_size=64, verbose=2)
```

```python
loss, acc = model.evaluate(X_test, y_test)
```

```python
print('손실값:', round(loss, 3))
print('정확도:', round(acc, 3))
```

```python
y_pred = model.predict(X_test)
print(y_pred[0:3])
```

```python
y_pred_class = model.predict_classes(X_test)
print(y_pred_class[0:3])
```

```python
from keras.datasets import reuters

(X_train, y_train), (X_test, y_test) = reuters.load_data(num_words=10000, test_split=0.3)
```

```python
import numpy as np
np.unique(y_train)
```

```python
data = reuters.get_word_index()
data
```

```python
data3 = {}
for key, value in data.items():
    data3[value+3] = key
data3
```

```python
for idx, word in enumerate(('<pad', '<sos>', '<unk>')):
    data3[idx] = word
```

```python
for j in range(5):
    print(' '.join([data3[i] for i in X_train[j]]))
    print()
```

```python
from keras.preprocessing import sequence

X_train = sequence.pad_sequences(X_train, maxlen=100)
X_test = sequence.pad_sequences(X_test, maxlen=100)

from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train) # 훈련용 뉴스 기사 레이블의 원-핫 인코딩
y_test = to_categorical(y_test) 
```

```python
for j in range(5):
    print(y_train[j])
```

```python
from keras.layers.embeddings import Embedding
from keras.models import Sequential
from keras.layers import LSTM, Dense

model = Sequential()
model.add(Embedding(input_dim = 10000,
                    output_dim = 32,
                    input_length = 100)) # X_train과 같이 100으로 맞춰야함
model.add(LSTM(units=100))
model.add(Dense(units=46, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train,
          epochs=10, batch_size=100, verbose=2)
```

```python
loss, acc = model.evaluate(X_test, y_test)
```

```python
print('손실값:', round(loss, 3))
print('정확도:', round(acc, 3))
```

```python
y_pred = model.predict(X_test)
print(y_pred[0:3])
```

```python
y_pred_class = model.predict_classes(X_test)
print(y_pred_class[0:3])
```
