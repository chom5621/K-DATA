```python
tr =[['우유', '계란', '커피'],
     ['빵', '우유', '버터', '휴지', '라면'],
     ['맥주','땅콩'],
     ['계란', '빵', '우유', '생수'],
     ['햇반', '커피', '간장', '계란']]
```

```python
import apyori as ap

result = ap.apriori(tr, min_support=0.3,
                   min_confidence=0.6)
rules = list(result)
```

```python
for r in rules:
    print(r, '\n') #빈발 항목규칙 5개
```

```python
r = list(rules[2])

print("빈발항목집합:", r[0])
print("지지도:", r[1])
print("연관규칙 정보:", r[2])
print("연관규칙의 개수:", len(r[2]))
```

```python
s = list(rules[2][2]) # 또는 s = list(r[2])

for i in s:
    print(i, '\n')
```

```python
p = list(rules[2][2][0]) # 또는 p = list(s[0])
                         # 또는 p = list(r[2][0])
print("선행절:",p[0])
print("후행절:",p[1])
print("신뢰도:",p[2])
print("향상도:",p[3])
```

```python
tr_num = 50

tr_len = list(range(2, 6))

item_domain = list(range(1, 21))
```

```python
data = []

for i in range(tr_num):
    current_tr = random.sample(item_domain, random.choice(tr_len))
    current_tr.sort()
    data.append(current_tr)
```

```python
data
```

```python
import apyori as ap

rules = list(ap.apriori(data, min_support=0.1, min_confidence=0.1))
```

```python
rules
```

```python
for r in rules:
    if len(r[0]) > 1:
        print("항목집합 =",r[0])
        print("지지도 =",r[1])
        for ar in list(r[2]):
            if len(ar[0]) >= 1 and len(ar[1]) >= 1:
                print(set(ar[0]),"=>",set(ar[1]))
                print("\t신뢰도 =", ar[2])
                print("\t향상도 =", ar[3])
        print()
```

```python
f = open("market.csv") 

tr = list(ap.load_transactions(f, delimiter=','))

rules = list(ap.apriori(tr, min_support=0.01, min_confidence=0.1))
```

```python
result = []
for r in rules:
    if len(r[0]) > 1:
        support = r[1]
        for ar in list(r[2]):
            row = [set(ar[0]), '=>', set(ar[1]), support, ar[2], ar[3]]
            result.append(row)

import pandas as pd
col = ['선행', '=>', '후행', '지지', '신뢰', '향상']
df = pd.DataFrame(result, columns=col)
df.sort_values(by='향상', ascending=False)
```
